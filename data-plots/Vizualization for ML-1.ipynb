```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4, 5]
train_loss = [0.9, 0.6, 0.5, 0.4, 0.35]
val_loss = [1.0, 0.7, 0.6, 0.5, 0.45]

plt.plot(epochs,train_loss, label="Training loss" )
#plt.show() #sperate graph/plot required
plt.title(" loss v/s echoch")
plt.grid(True)
plt.plot(epochs,val_loss,  label="Validation loss" )
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
```

Historgram - show the statistical distribution of feature

```python
import seaborn as sns
df = sns.load_dataset("iris")
plt.hist(df['petal_length'],bins=20) # bins means bricks/building split more granular distribution
plt.xlabel("length(cm)")
plt.ylabel("frequency")
```

Scatter Plot- 

```python
plt.scatter(df['sepal_length'], df['petal_length'], c=df['species'].astype('category').cat.codes)
```

Count Plot

```python
import  seaborn  as sns
df = sns.load_dataset('titanic')
df.head()
sns.countplot(x="survived", data =df )
plt.xlabel("survived(0=No, 1=Yes)")

```

Box Plot - Nurmerical feature vs Target

```python
import  seaborn  as sns
sns.boxplot(x='survived', y = 'age', data=df ,color='seagreen')
#sns.boxplot(x='age', y='class', data = df)
```

```python

sns.violinplot(x='pclass', y='age', hue='survived', split=True, data=df)
plt.title('Fare Distribution by class and survived')
plt.show()          
```

HeatMap - Correlation Matrix

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, linecolor='white', annot=True,cmap='coolwarm' )
plt.title('Feature Correlation')
plt.show()
```

PairPlot - Multivariate Analysis - compare features pair-wise

```python
sns.pairplot(df[['age','fare', 'pclass', 'survived']], hue='survived')
plt.suptitle('pair plot of features')
plt.show()
```

Plotly Interactive Visual - hovering mouse gets/shows the intuitive data

```python
import plotly.express as ex
fig = ex.scatter(df, x='age', y='fare', color='survived', size='pclass' ,
 hover_data=['sex', 'embarked', 'alone'], title='Interactive Titanic Plot')

fig.show()

```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", size="petal_length")
fig.show()
```

3D Scatter Plot 

```python
import plotly.express as px

df = px.data.iris()

fig = px.scatter_3d(
    df,
    x="sepal_width",
    y="sepal_length",
    z="petal_width",
    color="species",
    size="petal_length",
    symbol="species",
    opacity=0.8
)

fig.show()
```
